# LINKED LIST

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None


class linkedList:
    def __init__(self):
        self.head = None

    def push(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def append(self, data):
        temp = self.head
        while temp.next:
            temp = temp.next
        new_node = Node(data)
        temp.next = new_node

    def add_current(self, data, index):
        node = Node(data)
        temp = self.head
        if index == 0:
            node.next = self.head
            self.head = node
        else:
            for i in range(index - 1):
                temp = temp.next
            node.next = temp.next
            temp.next = node

    def add_after(self, data, loc):
        temp = self.head
        while temp:
            if temp.data == loc:
                break
            else:
                temp = temp.next
        new_node = Node(data)
        new_node.next = temp.next
        temp.next = new_node

    def add_before(self, data, loc):
        temp = self.head
        while temp.next:
            if temp.next.data == loc:
                break
            else:
                temp = temp.next
        new_node = Node(data)
        new_node.next = temp.next
        temp.next = new_node

    def print(self):
        temp = self.head
        while temp is not None:
            print(temp.data, '-->', end='')
            temp = temp.next
        print(" ")

    def remove_beg(self):
        self.head = self.head.next

    def remove_end(self):
        temp = self.head
        while temp.next.next:
            temp = temp.next
        temp.next = None

    def remove_after(self, loc):
        temp = self.head
        while temp:
            if temp.data == loc:
                break
            else:
                temp = temp.next
        temp.next = temp.next.next

    def remove_before(self, loc):
        temp = self.head
        while temp.next.next:
            if temp.next.next.data == loc:
                break
            else:
                temp = temp.next
        temp.next = temp.next.next

    def remove_current(self, loc):
        temp = self.head
        while temp.next:
            if temp.next.data == loc:
                break
            else:
                temp = temp.next
        temp.next = temp.next.next


l = linkedList()
l.push(1)
l.push(2)
l.push(3)
l.push(4)
l.append(5)
l.add_after(6, 5)
l.add_after(20, 3)
l.add_after(30, 3)
l.add_before(5000, 30)
l.remove_beg()
l.remove_end()
l.remove_after(30)
l.remove_before(30)
l.remove_current(30)
l.add_current(300, 0)
l.print()
```
